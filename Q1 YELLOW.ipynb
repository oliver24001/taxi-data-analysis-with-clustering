```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('yellow2019_0.csv', parse_dates= ['tpep_pickup_datetime','tpep_dropoff_datetime'])
data
```

```python
data['travel_time'] = data['dropoff_datetime']-data['pickup_datetime']
data
```

```python
data.to_csv
```

```python
data.groupby(data['tpep_pickup_datetime'].dt.hour).count()
```

```python
pickupdf=data.sort_values(by=['tpep_pickup_datetime'])
pickupdf
```

```python
pickup_byhour=data.groupby([data['tpep_pickup_datetime'].dt.hour,data['PULocationID']]).size()
pickup_byhour
pickup_byhour=pd.DataFrame(pickup_byhour)
pickup_byhour.to_csv('YELLOW_pickup_byhour.csv')

```

```python
pickup_byday=data.groupby([data['tpep_pickup_datetime'].dt.day,data['PULocationID']]).size()
pickup_byday
pickup_byday=pd.DataFrame(pickup_byday)
pickup_byday.to_csv('YELLOW_pickup_byday.csv')
```

```python
pickuppeak_time=data.groupby([data['tpep_pickup_datetime'].dt.hour]).size()
pickuppeak_time
```

```python
dropoffdf=data.sort_values(by=['tpep_dropoff_datetime'])
dropoffdf
```

```python
dropoff_byhour=data.groupby([data['tpep_dropoff_datetime'].dt.hour,data['DOLocationID']]).size()
dropoff_byhour
dropoff_byhour=pd.DataFrame(dropoff_byhour)
dropoff_byhour.to_csv('YELLOW_dropoff_byhour.csv')
```

```python
dropoff_byday=data.groupby([data['tpep_dropoff_datetime'].dt.day,data['DOLocationID']]).size()
dropoff_byday
dropoff_byday=pd.DataFrame(dropoff_byday)
dropoff_byday.to_csv('YELLOW_dropoff_byday.csv')
```
