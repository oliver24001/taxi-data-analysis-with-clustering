```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('HVFHV2019_0_withzoning.csv')
data = pd.read_csv('HVFHV2019_0_withzoning.csv', parse_dates= ['pickup_datetime','dropoff_datetime'])
data
```

```python
data['travel_time'] = data['dropoff_datetime']-data['pickup_datetime']
data
```

```python
data.to_csv
```

```python
data.groupby(data['pickup_datetime'].dt.hour).count()
```

```python
pickupdf=data.sort_values(by=['pickup_datetime'])
pickupdf
```

```python
pickup_byhour=data.groupby([data['pickup_datetime'].dt.hour,data['PULocationID']]).size()
pickup_byhour
pickup_byhour=pd.DataFrame(pickup_byhour)
pickup_byhour.to_csv('pickup_byhour.csv')

```

```python
pickup_byday=data.groupby([data['pickup_datetime'].dt.day,data['PULocationID']]).size()
pickup_byday
pickup_byday=pd.DataFrame(pickup_byday)
pickup_byday.to_csv('pickup_byday.csv')
```

```python
pickuppeak_time=data.groupby([data['pickup_datetime'].dt.hour]).size()
pickuppeak_time
```
